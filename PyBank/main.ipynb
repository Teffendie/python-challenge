```python
import os
import csv

pybank_csv = 'Resources/budget_data.csv'

def totalMonths(filename):
    with open(filename) as f:
        csvreader = csv.reader(f, delimiter=",")
        next(csvreader, None)
        return (sum(1 for line in csvreader))



#------------------------------------------------------------------------------
def totalAmount(filename):
    with open(filename) as f:
        csvreader = csv.reader(f, delimiter=",")
        next(csvreader, None)
        return (sum(int(row[1]) for row in csvreader))
    


#------------------------------------------------------------------------------
def averageChange(filename):
    with open(filename) as f:
        csvreader = csv.reader(f, delimiter=",")
        next(csvreader, None)
        profitlosses = []
        
        for row in csvreader:
            profitloss = row[1]
            profitlosses.append(profitloss)        
        
        return (sum((int(profitlosses[i+1]) - int(profitlosses[i]) for i in range(len(profitlosses)-1)))/(totalMonths(filename)-1))    
          

#------------------------------------------------------------------------------
def greatestLowestIncrease(filename):
    with open(filename) as f:
        csvreader = csv.reader(f, delimiter=",")
        next(csvreader, None)
        dates = []
        profitlosses = []
        
        for row in csvreader:
            date = row[0]
            profitloss = row[1]
            dates.append(date)
            profitlosses.append(profitloss)  
              
            v =[int(profitlosses[i+1]) - int(profitlosses[i]) for i in range(len(profitlosses)-1)]   
        
        highest = max(v)
        highestIndexNumber = v.index(highest)
        highestdate = dates[highestIndexNumber + 1]
        lowest = min(v)
        lowestIndexNumber = v.index(lowest)
        lowestdate = dates[lowestIndexNumber + 1]        

        r = [highestdate,highest,lowestdate,lowest]
        
        return(r)
    
#========================================================================

d = greatestLowestIncrease(pybank_csv)

p = """
Financial Analysis
----------------------------
"""
print(p)
print(f'Total Months: {totalMonths(pybank_csv)}')
print(f'Total: {totalAmount(pybank_csv)}')
print(f'Average  Change: {round(averageChange(pybank_csv), 2)}')
print(f'Greatest Increase in Profits: {d[0]} (${d[1]})')
print(f'Greatest Decrease in Profits: {d[2]} (${d[3]})')



with open("Output.txt", "w") as text_file:
    line1 = """
Financial Analysis
----------------------------
    """
    line2 = f'Total Months: {totalMonths(pybank_csv)}'
    line3 = f'Total: {totalAmount(pybank_csv)}'
    line4 = f'Average  Change: {round(averageChange(pybank_csv), 2)}'
    line5 = f'Greatest Increase in Profits: {d[0]} (${d[1]})'
    line6 = f'Greatest Decrease in Profits: {d[2]} (${d[3]})'
    text_file.write('{}\n{}\n{}\n{}\n{}\n{}\n'.format(line1,line2,line3,line4,line5,line6))
```
